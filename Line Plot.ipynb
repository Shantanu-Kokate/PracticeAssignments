## Line Plot for N-N, N-C, C-C

```python
import seaborn as sns             # for data visualization
import pandas as pd               # for data analysis
import matplotlib.pyplot as plt   # for data visualization
```

```python
df = pd.read_csv('tip_line.csv')[:100:5]    #slicing is done to get different categories and less data for analysis
df
```

```python
df.shape
```

### Line Plot : Using Seaborn

```python
## you can use following legends
# sns.set(style='darkgrid',)
# plt.title("Line Plot", fontsize = 20)       # for title
# plt.xlabel("x-label information", fontsize = 15)  # label for x-axis
# plt.ylabel("y-label information", fontsize = 15)  # label for y-axis
```

```python
# plotting simple bar plot for N-N category

sns.set(style='darkgrid',)
plt.figure(figsize=(6,3))
sns.lineplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.lineplot(x = "tip", y = "total_bill", data = df, hue = "smoker")
plt.show()
```

```python
# plotting two categories of same feature in the single plot

plt.figure(figsize=(6,3))
sns.lineplot(x = "size", y = "total_bill", data = df, hue = "sex", markers = ["o", "<", ">", "^"], dashes=False)
plt.show()
```

```python
style => Give style to line plot, like dashes. Different for each line plot.
palette => Give colormap for graph. You can choose anyone from bellow which is separated by a comma.
dashes => If line plot with dashes then use “False” value for no dashes otherwise “True“.
markers => Give the markers for point like (x1,y1). for markers follow matplotlib line plot blog.
legend => Give legend. The default value is “brief” but you can give “full” or “False“. False for no legend.
```

### Line Plot Using Matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
# Drawing simple line chart

plt.figure(figsize=(5,3))
plt.plot('total_bill', 'tip',data=df)

plt.title("Tip amount on Total Bill")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```

```python
# Changing the X and Y-axis values (zoom in/out the graph)

plt.figure(figsize=(5,3))
plt.plot('total_bill', 'tip',data=df)
plt.axis([0,150, 0,12])     # sets the axis value: plt.axis(xmin, xmax, ymin, ymax)

plt.title("Tip amount on Total Bill")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```

```python
# Changing the X and Y-axis values (zoom in/out the graph)

plt.figure(figsize=(5,3))
plt.plot('total_bill', 'tip',data=df)
plt.axis([0,30, 0,6])                # sets the axis value: plt.axis(xmin, xmax, ymin, ymax)

plt.title("Tip amount on Total Bill")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```

```python
# applying style to graph

plt.figure(figsize=(5,3))
plt.plot('total_bill', 'tip',data=df, color='g', linestyle='--', marker='o')

plt.title("Tip amount on Total Bill")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```
